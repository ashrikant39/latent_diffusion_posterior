```python
image_path = "images/bedrooms/resized.png"
train_snr = 20
test_snr = 15
base_dir = "/nfs/turbo/coe-hunseok/ashri/lsun_data/loggings/jscc_models_updated"
LOGDIR = glob.glob(os.path.join(base_dir, f"snr_{train_snr}", "*"))[0]
checkpoint_path = os.path.join(LOGDIR, "checkpoints", "last.ckpt")
config = OmegaConf.load(glob.glob(os.path.join(LOGDIR, "configs", "*project.yaml"))[0])
save_image_path = f"images/bedrooms/snr_0/saved_image_train_{train_snr}_test_{test_snr}.png"
```

```python
model, _ = load_model(config, checkpoint_path, gpu = True, eval_mode = True)
```

```python
scale_function = lambda img: (2*img - 1.0)
ScaleShiftTransform = torchvision.transforms.Lambda(scale_function)
data_transform = Compose(
        [
            Resize(size = (256, 256)),
            ToTensor(),
            ScaleShiftTransform
        ]
    )
```

```python
image = Image.open(image_path)
image = data_transform(image)[None].cuda()
image.shape
```

```python
reconstructed_image = deep_jscc_testing(model, image, test_snr)
```

```python
recon_image = custom_to_pil(reconstructed_image[0])
recon_image.save(save_image_path)
```

```python
PSNR = psnr(image, reconstructed_image)
PSNR
```
