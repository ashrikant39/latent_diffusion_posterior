```python
import pickle
import glob
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
snr_list = range(-5, 30, 5)
```

```python
metric_list = ["PSNR", "MS_SSIM", "FID", "LPIPS_VGG"]
metric_names = ["Peak Signal to Noise Ratio (dB)", "Multi Scale SSIM", "FID Score", "LPIPS Score (AlexNet)", "LPIPS Score (VGG)", "LPIPS Score (SqeezeNet)"]
```

```python
denoised_metrics = {}
noisy_metrics = {}
```

```python
for metric in metric_list:

    denoised_metrics[metric] = []
    noisy_metrics[metric] = []

    for snr in snr_list:

        denoised_filepath = f"scores/denoised_scores_{snr+100}_dict.pkl"
        noisy_filepath = f"scores/noisy_latent_scores_{snr+100}_dict.pkl"
        jscc_filepath = f"/nfs/turbo/coe-hunseok/ashri/lsun_data/testing_results/bedrooms/jscc_scores_{int(SNR_DB)+100}_dict.pkl"

        denoised_snr_metric = pickle.load(open(denoised_filepath, "rb"))
        noisy_snr_metric = pickle.load(open(noisy_filepath, "rb"))

        denoised_metrics[metric].append(denoised_snr_metric[metric])
        noisy_metrics[metric].append(noisy_snr_metric[metric])
```

```python
metric_list[0]+"dB"
```

```python
for metric, name in zip(metric_list, metric_names):

    metric_short = (metric + "(dB)") if metric=="PSNR" else metric
    figname = f"scores/denoised_plot_{metric}.png"
    plt.figure(figsize=(10, 4))
    plt.plot(snr_list, denoised_metrics[metric], color = 'b', label = "With Latent Denoising")
    plt.plot(snr_list, noisy_metrics[metric], linestyle = "dashed", color='r', label = "Without Latent Denoising")
    
    
    plt.title(f"{name} vs Channel-SNR")
    plt.xlabel("Channel SNR (dB)")
    plt.ylabel(f"{metric_short}")
    plt.legend()
    plt.savefig(figname)
    plt.show()
```

```python
import pandas as pd
```

```python
save_denoised = denoised_metrics.copy()
save_noisy = noisy_metrics.copy()
save_denoised["SNR"] = snr_list
save_noisy["SNR"] = snr_list
```

```python
denoised_df = pd.DataFrame.from_dict(save_denoised)
denoised_df = denoised_df[["SNR", *metric_list]]
denoised_df
```

```python
noisy_df = pd.DataFrame.from_dict(save_noisy)
noisy_df = noisy_df[["SNR", *metric_list]]
noisy_df
```

```python
denoised_csv_path = "scores/denoised.csv"
noisy_csv_path = "scores/noisy.csv"
denoised_df.to_csv(denoised_csv_path)
noisy_df.to_csv(noisy_csv_path)
```
